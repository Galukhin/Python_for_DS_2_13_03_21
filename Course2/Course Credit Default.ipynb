**Задача**

Требуется, на основании имеющихся данных о клиентах банка, построить модель, используя обучающий датасет, для прогнозирования невыполнения долговых обязательств по текущему кредиту. Выполнить прогноз для примеров из тестового датасета.

**Целевая переменная**

Credit Default - факт невыполнения кредитных обязательств

**Метрика качества**

* F1-score (sklearn.metrics.f1_score)
* F1 > 0.5
* Метрика оценивается по качеству прогноза для главного класса (1 - просрочка по кредиту)

### Класс предобратки данных
Сведем всю необходимую предобратку в один класс. Заполнять все пропуски и выбросы будем с помощью моделей GradientBoostingRegressor с параметрами по умолчанию. Для оценки качества заполнения моделями также визуализируем предсказания на тестовой и валидной выборке. Пропуски в признаке Years in current job заполним модой(признак не заполняется моделью). Признак Months since last delinquent, как уже писал выше, заполним -1, моделью признак также не заполняется.

```python
class DataPreparation:
    """Подготовка исходных данных"""

    def __init__(self):
        """Параметры класса"""
        self.medians = None
        self.YCJ_mode = None
        self.MD_qntl_l = None
        self.models_dict = {}
        self.outliers_dict = {'Maximum Open Credit': 'MOC_outlier', 
                              'Monthly Debt': 'MD_outlier',
                              'Current Loan Amount': 'CLA_outlier',
                              'Credit Score': 'CS_outlier',
                              'Annual Income': 'AI_outlier', 
                              'Bankruptcies': 'B_outlier'}
        self.outliers_float_dict = {'Maximum Open Credit': 'MOC_outlier',
                               'Monthly Debt': 'MD_outlier',
                               'Current Loan Amount': 'CLA_outlier',
                               'Credit Score': 'CS_outlier',
                               'Annual Income': 'AI_outlier'}
        self.outliers_int_dict = {'Bankruptcies': 'B_outlier'}
        
        
    def fit(self, X):
        """Сохранение статистик"""       
        # Расчет медиан
        self.medians = X.median()
        self.YCJ_mode = X['Years in current job'].mode()[0]
        self.MD_qntl_l = X['Monthly Debt'].quantile(.025)
                
        
    def pretransform(self, X):
        """Трансформация данных"""
        
        # Home Ownership
        X.loc[X['Home Ownership'] == 'Have Mortgage', 'Home Ownership'] = 'Home Mortgage'
        
        # Annual Income
        cond_AI = X['Annual Income'].isna()
        X['AI_outlier'] = 0
        X.loc[cond_AI, 'AI_outlier'] = 1
        
        # Years in current job
        cond_YCJ = X['Years in current job'].isna()
        X['YCJ_outlier'] = 0
        X.loc[cond_YCJ, 'YCJ_outlier'] = 1
        X['Years in current job'].fillna(self.YCJ_mode, inplace=True)
        
        # Tax liens
        X.loc[X['Tax Liens'] == 1, 'Tax Liens'] = 0
        X.loc[X['Tax Liens'] > 1, 'Tax Liens'] = 1
        
        # Maximum open credit
        cond_MOC = (X['Maximum Open Credit'] > 1e7) | X['Maximum Open Credit'].isna()
        X['MOC_outlier'] = 0
        X.loc[cond_MOC, 'MOC_outlier'] = 1
        
        # Number of Credit Problems
        X.loc[X['Number of Credit Problems'] > 4, 'Number of Credit Problems'] = 4
        
        # Months since last delinquent
        cond_MSLD = X['Months since last delinquent'].isna()
        X['MSLD_outlier'] = 0
        X.loc[cond_MSLD, 'MSLD_outlier'] = 1
        X['Months since last delinquent'].fillna(-1, inplace=True)
        
        # Bankruptcies
        X.loc[X['Bankruptcies'] > 2, 'Bankruptcies'] = 2
        cond_B = X['Bankruptcies'].isna()
        X['B_outlier'] = 0
        X.loc[cond_B, 'B_outlier'] = 1
        
        # Purpose
        X.loc[X['Purpose'] == 'renewable energy', 'Purpose'] = 'small business'
        
        # Current Loan Amount
        CLA_max = 1e6
        cond_CLA = (X['Current Loan Amount'] > CLA_max) | X['Current Loan Amount'].isna()
        X['CLA_outlier'] = 0
        X.loc[cond_CLA, 'CLA_outlier'] = 1
        
        # Monthly Debt
        cond_MD = (X['Monthly Debt'] < self.MD_qntl_l) | X['Monthly Debt'].isna()
        X['MD_outlier'] = 0
        X.loc[cond_MD, 'MD_outlier'] = 1
        
        # Credit Score
        X.loc[X['Credit Score'] > 1000, 'Credit Score'] = X.loc[X['Credit Score'] > 1000, 'Credit Score'] / 10
        cond_CS = X['Credit Score'].isna()
        X['CS_outlier'] = 0
        X.loc[cond_CS, 'CS_outlier'] = 1
        
        return X
        
    def fit_futures(self, X):
        """
        Обучение моделей для заполнения фичей
        """
        df_copy = X.copy()
        df_copy = self.transform_for_fit_futures(df_copy)
        
        df_copy = df_copy.loc[df_copy['sum_outliers'] == 0]
        df_copy = df_copy.drop(columns=self.outliers_dict.values())
        df_copy = df_copy.drop(columns='sum_outliers')
        df_copy = df_copy.drop(columns='Months since last delinquent')
        df_copy = df_copy.drop(columns='MSLD_outlier')
        # Следующие два цикла - для визуализации и оценки качества моделей
        for itm in self.outliers_float_dict.keys():
    
            X = df_copy.drop(columns=itm)
            y = df_copy[itm]
            X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
    
            gbr = GradientBoostingRegressor()
            gbr.fit(X_train, y_train)

            y_train_preds = gbr.predict(X_train)
            y_test_preds = gbr.predict(X_valid)

            self.evaluate_preds(y_train, y_train_preds, y_valid, y_test_preds, save=True, prefix=itm)
        
        for itm in self.outliers_int_dict.keys():
    
            X = df_copy.drop(columns=itm)
            y = df_copy[itm]
            X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
    
            gbr = GradientBoostingRegressor()
            gbr.fit(X_train, y_train)

            y_train_preds = pd.Series(gbr.predict(X_train), index=y_train.index)
            y_train_preds = y_train_preds.map(round).map(int).clip(0)
            y_test_preds = pd.Series(gbr.predict(X_valid), index=y_valid.index)
            y_test_preds = y_test_preds.map(round).map(int).clip(0)

            self.evaluate_preds(y_train, y_train_preds, y_valid, y_test_preds, save=True, prefix=itm)
        # Следующие два цикла - для обучения моделей
        self.models_dict = {}
        for itm in self.outliers_float_dict.keys():
    
            X = df_copy.drop(columns=itm)
            y = df_copy[itm]
            
            self.models_dict.update({itm: [None, None]})
            self.models_dict[itm][0] = GradientBoostingRegressor()
            self.models_dict[itm][1] = X.columns
            self.models_dict[itm][0].fit(X, y)
        
        for itm in self.outliers_int_dict.keys():
    
            X = df_copy.drop(columns=itm)
            y = df_copy[itm]
            
            self.models_dict.update({itm: [None, None]})
            self.models_dict[itm][0] = GradientBoostingRegressor()
            self.models_dict[itm][1] = X.columns
            self.models_dict[itm][0].fit(X, y)
    
    def transform(self, X):
        """
        Заполнение пропусков и выбросов на train и test с помощью моделей из self.models_dict
        и уменьшение занимаемой памяти датасета
        """
        
        df_copy = X.copy()
        df_copy = self.transform_for_fit_futures(df_copy)
        
        for itm in self.outliers_float_dict.keys():
            X_outlier = df_copy.loc[df_copy[self.outliers_float_dict[itm]] == 1, self.models_dict[itm][1]]
            y_preds = self.models_dict[itm][0].predict(X_outlier)
            df_copy.loc[df_copy[self.outliers_float_dict[itm]] == 1, itm] = y_preds
            X.loc[X[self.outliers_float_dict[itm]] == 1, itm] = y_preds
        
        for itm in self.outliers_int_dict.keys():
            X_outlier = df_copy.loc[df_copy[self.outliers_int_dict[itm]] == 1, self.models_dict[itm][1]]
            y_preds = pd.Series(self.models_dict[itm][0].predict(X_outlier), index=X_outlier.index)
            y_preds = y_preds.map(round).map(int).clip(0)
            df_copy.loc[df_copy[self.outliers_int_dict[itm]] == 1, itm] = y_preds
            X.loc[X[self.outliers_int_dict[itm]] == 1, itm] = y_preds
        
        X['Tax Liens'] = X['Tax Liens'].astype(np.int64)
        X['Number of Open Accounts'] = X['Number of Open Accounts'].astype(np.int64)
        X['Years of Credit History'] = X['Years of Credit History'].astype(np.int64)
        X['Number of Credit Problems'] = X['Number of Credit Problems'].astype(np.int64)
        X['Bankruptcies'] = X['Bankruptcies'].astype(np.int64)
        
        X = self.reduce_mem_usage(X)
        
        return X
    
    def transform_for_fit_futures(self, df):
        """
        Преобразование категорий и уменьшение объема занимаемой памяти
        перед заполнением пропусков и выбросов с помощью моделей
        """
        
        for itm in self.outliers_dict.keys():
            df.loc[df[self.outliers_dict[itm]] == 1, itm] = self.medians[itm]
        
        df = pd.concat([df, pd.get_dummies(df['Home Ownership'])], axis=1)
        df = df.drop(columns='Home Ownership')
        
        order = ["< 1 year", "1 year", "2 years", "3 years", "4 years", "5 years", 
                 "6 years", "7 years", "8 years", "9 years", "10+ years"]
        for itm in order:
            df.loc[df['Years in current job'] == itm, 'Years in current job'] = str(order.index(itm))
        df['Years in current job'] = df['Years in current job'].astype(np.float64)
        
        df = pd.concat([df, pd.get_dummies(df['Purpose'])], axis=1)
        df = df.drop(columns='Purpose')
        
        df = pd.concat([df, pd.get_dummies(df['Term'])], axis=1)
        df = df.drop(columns='Term')
        
        outliers = ['AI_outlier', 'MOC_outlier', 'CLA_outlier', 
                    'MD_outlier', 'CS_outlier', 'B_outlier']
        df['sum_outliers'] = 0
        for itm in outliers:
            df['sum_outliers'] = df['sum_outliers'] + df[itm]
        
        df['Tax Liens'] = df['Tax Liens'].astype(np.int64)
        df['Number of Open Accounts'] = df['Number of Open Accounts'].astype(np.int64)
        df['Years of Credit History'] = df['Years of Credit History'].astype(np.int64)
        df['Number of Credit Problems'] = df['Number of Credit Problems'].astype(np.int64)
        df['Bankruptcies'] = df['Bankruptcies'].astype(np.int64)
        
        df = self.reduce_mem_usage(df)
        
        return df

    @staticmethod
    def reduce_mem_usage(df):
        """
        iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.
        """        
           
        for col in df.columns:
            col_type = df[col].dtype
        
            if col_type != object:
                c_min = df[col].min()
                c_max = df[col].max()
                if str(col_type)[:3] == 'int':
                    if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                        df[col] = df[col].astype(np.int8)
                    elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                        df[col] = df[col].astype(np.int16)
                    elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                        df[col] = df[col].astype(np.int32)
                    elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                        df[col] = df[col].astype(np.int64)
                elif str(col_type)[:4] == 'uint':
                    if c_min > np.iinfo(np.uint8).min and c_max < np.iinfo(np.uint8).max:
                        df[col] = df[col].astype(np.uint8)
                    elif c_min > np.iinfo(np.uint16).min and c_max < np.iinfo(np.uint16).max:
                        df[col] = df[col].astype(np.uint16)
                    elif c_min > np.iinfo(np.uint32).min and c_max < np.iinfo(np.uint32).max:
                        df[col] = df[col].astype(np.uint32)
                    elif c_min > np.iinfo(np.uint64).min and c_max < np.iinfo(np.uint64).max:
                        df[col] = df[col].astype(np.uint64)
                else:
                    if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                        df[col] = df[col].astype(np.float32)
                    else:
                        df[col] = df[col].astype(np.float64)
            else:
                df[col] = df[col].astype('category')
 
        return df

    @staticmethod
    def evaluate_preds(train_true_values, train_pred_values, test_true_values, test_pred_values, save=False, prefix=None):
    
        plt.figure(figsize=(18,10))
    
        train_min = train_true_values.min()
        train_max = train_true_values.max()
    
        if prefix is not None:
            train_title = 'Train sample prediction ' + prefix + '\n' + 'Train R2: ' + str(round(r2(train_true_values, train_pred_values), 3))
            test_title = 'Test sample prediction ' + prefix + '\n' + 'Test R2: ' + str(round(r2(test_true_values, test_pred_values), 3))
        else:
            train_title = 'Train sample prediction ' + '\n' + 'Train R2: ' + str(round(r2(train_true_values, train_pred_values), 3))
            test_title = 'Test sample prediction ' + '\n' + 'Test R2: ' + str(round(r2(test_true_values, test_pred_values), 3))
    
        plt.subplot(121)
        sns.scatterplot(x=train_pred_values, y=train_true_values, cmap=plt.get_cmap('jet'), alpha=0.5)
        plt.plot([train_min, train_max], [train_min, train_max], linestyle='--', color='black')  # диагональ, где true_values = pred_values
        plt.xlabel('Predicted values')
        plt.ylabel('True values')
        plt.title(train_title)
    
        plt.subplot(122)
        sns.scatterplot(x=test_pred_values, y=test_true_values, cmap=plt.get_cmap('jet'), alpha=0.5)
        plt.plot([train_min, train_max], [train_min, train_max], linestyle='--', color='black')  # диагональ, где true_values = pred_values
        plt.xlabel('Predicted values')
        plt.ylabel('True values')
        plt.title(test_title)

        if save == True:
            plt.savefig('./reports/' + prefix + '_report.png')
```

```python
X = df.drop(columns='Credit Default')
y = df['Credit Default']
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, 
                                                      shuffle=True, 
                                                      stratify=y, 
                                                      random_state=42)
```

```python
prepare = DataPreparation()

prepare.fit(X_train)
```

```python
prepare.fit_futures(X_train)
```

У признаков Annual Income и Credit Score достаточно низкая метрика качества, но это все равно лучше, чем заполнять медианой или средним

```python
X_train = prepare.transform(X_train)
X_valid = prepare.transform(X_valid)
```

```python
X_train.isna().sum().sum(), X_valid.isna().sum().sum()
```

```python
def get_classification_report(y_train_true, y_train_pred, y_test_true, y_test_pred):
    print('TRAIN\n\n' + classification_report(y_train_true, y_train_pred))
    print('TEST\n\n' + classification_report(y_test_true, y_test_pred))
    print('CONFUSION MATRIX\n')
    print(pd.crosstab(y_test_true, y_test_pred))
```

```python
def evaluate_preds(model, X_train, X_test, y_train, y_test):
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    get_classification_report(y_train, y_train_pred, y_test, y_test_pred)
```

```python
CAT_FEATURE_NAMES = ['Home Ownership', 'Years in current job', 'Tax Liens', 'Number of Open Accounts',
                     'Bankruptcies', 'Purpose', 'Term']
```

```python
disbalance = y_train.value_counts()[0] / y_train.value_counts()[1]
disbalance
```

```python
frozen_params = {
     'class_weights':[1, disbalance], 
     'silent':True,
     'random_state':21,
     'cat_features':CAT_FEATURE_NAMES,
     'one_hot_max_size':15,
     'eval_metric':'F1',
     'early_stopping_rounds':20
}
```

Поиск по сетке для CatBoost дал следующие параметры: 'depth': 3, 'iterations': 50

```python
%%time

final_model = catb.CatBoostClassifier(**frozen_params, iterations=50, max_depth=3)
final_model.fit(X_train, y_train, plot=True, eval_set=(X_valid, y_valid))

evaluate_preds(final_model, X_train, X_valid, y_train, y_valid)
```

```python
feature_importances = pd.DataFrame(zip(X_train.columns, final_model.feature_importances_), 
                                   columns=['feature_name', 'importance'])

feature_importances.sort_values(by='importance', ascending=False)

plt.figure(figsize = (20, len(feature_importances) * 0.355))
    
sns.barplot(feature_importances['importance'], feature_importances['feature_name'])
    
plt.xlabel('Importance')
plt.title('Importance of features')
plt.show()
```
